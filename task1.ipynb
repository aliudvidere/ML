```python
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/"
data = pd.read_csv(DATA_URL + "adult.data.csv")
data.head()
```

1. How many men and women (sex feature) are represented in this dataset?

```python
data.groupby(data["sex"]).count()
```

```python
d = {"<=50K" : False, ">50K" : True}
data.replace({"salary": d}, inplace=True)
```

```python
data["salary > 50K"] = data["salary"]
```

```python
data.drop("salary", axis=1, inplace=True)
```

```python
data["female"] = data["sex"]
data.drop("sex", axis=1, inplace=True)
```

```python
sex_d = {"Female" : True, "Male" : False}
data["female"] = data["female"].map(sex_d)
```

8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

```python
married_list = ["Married-civ-spouse", "Married-spouse-absent", "Married-AF-spouse"]
data["married"] = data["marital-status"].apply(lambda x : True if x in married_list else False)
```

```python
data.groupby("married")["salary > 50K"].value_counts(normalize=True)
```

9. What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

```python
max_hour_per_week = data["hours-per-week"].max()
```

```python
data[data["hours-per-week"] == max_hour_per_week]["salary > 50K"].value_counts(normalize=True)
```
